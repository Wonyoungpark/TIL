```python
import numpy as np
import matplotlib.pyplot as plt

fruits = np.load('fruits_300.npy')
print(fruits.shape)
```

```python
print(fruits[0,0,:])
```

```python
plt.imshow(fruits[0],cmap='gray') # 바탕과 물체 색을 반전
plt.show() # 사과
```

```python
plt.imshow(fruits[0],cmap='gray_r') # 바탕을 다시 하얗게 만듦
plt.show()
```

```python
fig, axs = plt.subplots(1,2)
axs[0].imshow(fruits[100],cmap='gray_r') # 파인애플
axs[1].imshow(fruits[200],cmap='gray_r') # 바나나
plt.show()
```

```python
apple = fruits[0:100].reshape(-1,100*100)
pineapple = fruits[100:200].reshape(-1,100*100)
banana = fruits[200:300].reshape(-1,100*100)
print(apple.shape)
```

```python
print(apple.mean(axis=1)) # 열 기준으로 각 사과별 평균 계산
```

```python
plt.hist(np.mean(apple,axis=1), alpha=0.8)
plt.hist(np.mean(pineapple,axis=1), alpha=0.8)
plt.hist(np.mean(banana,axis=1),alpha=0.8)
plt.legend(['apple','pineapple','banana'])
plt.show()
```

```python
# 전체 사과의 픽셀별 평균 계산
fig, axs = plt.subplots(1,3,figsize=(20,5))
axs[0].bar(range(10000), np.mean(apple,axis=0))
axs[1].bar(range(10000), np.mean(pineapple,axis=0))
axs[2].bar(range(10000), np.mean(banana,axis=0))
plt.show()
```

```python
apple_mean = np.mean(apple,axis=0).reshape(100,100)
pineapple_mean = np.mean(pineapple,axis=0).reshape(100,100)
banana_mean = np.mean(banana,axis=0).reshape(100,100)

fig,axs = plt.subplots(1,3,figsize=(20,5))
axs[0].imshow(apple_mean,cmap='gray_r')
axs[1].imshow(pineapple_mean,cmap='gray_r')
axs[2].imshow(banana_mean,cmap='gray_r')
plt.show()
```

```python
# 평균값에 가까운 사진 출력
abs_diff = np.abs(fruits - apple_mean) # 전체 - 사과 평균
abs_mean = np.mean(abs_diff, axis=(1,2))
print(abs_mean.shape)
```

```python
# [전체 - 사과 평균]이 작은 순서대로 나열 (=사과 평균과 거의 근접한 값 순서)
apple_index = np.argsort(abs_mean)[:100]

fig,axs = plt.subplots(10,10,figsize=(10,10))
for i in range(10):
  for j in range(10):
    axs[i,j].imshow(fruits[apple_index[i*10+j]], cmap='gray_r')
    axs[i,j].axis('off') # 좌표축 제외

plt.show()
```

```python
abs_diff_2 = np.abs(fruits - banana_mean)
abs_mean_2 = np.mean(abs_diff_2, axis=(1,2))
print(abs_mean_2.shape)
```

```python
banana_index = np.argsort(abs_mean_2)[:100]
fig,axs = plt.subplots(10,10,figsize=(10,10))
for i in range(10):
  for j in range(10):
    axs[i,j].imshow(fruits[banana_index[i*10+j]], cmap='gray_r')
    axs[i,j].axis('on')

plt.show()
```
